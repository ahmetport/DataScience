# Boxplot

**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**
    
**Her iki cinsiyette de ölüm şekline göre boxplot ile görselleştirin**

**Boxplot,** bir veri setinin özet istatistiklerini görselleştirmek için kullanılır. 

Boxplot, veri setindeki merkezi eğilim, yayılım ve çarpıklık gibi özellikleri anlamak için kullanışlıdır. 

Ayrıca, aykırı değerleri (outliers) belirlemek için de boxplot sıklıkla kullanılır. 

1. **Median (Medyan):**
   Kutunun ortasında çizili yatay bir çizgidir ve veri setinin orta değerini gösterir.
   
2. **Upper Quartile (Q3 - Üçüncü Çeyrek):**
   Kutunun üst kenarıdır ve veri setinin %75. persentil değerini gösterir.

3. **Lower Quartile (Q1 - Birinci Çeyrek):**
   Kutunun alt kenarıdır ve veri setinin %25. persentil değerini gösterir.

4. **Interquartile Range (IQR - Çeyrekler Açıklığı):**
   Q3 ile Q1 arasındaki farktır ve veri setinin çoğunu içerir.

5. **Whiskers:**
   Kutudan çıkan çizgilerdir ve genellikle veri setindeki minimum ve maksimum değerleri temsil eder. Ancak, bazı durumlarda aykırı değerlere kadar uzanmaz.

6. **Outliers (Aykırı Değerler):**
   Whisker'ların ötesindeki noktalardır ve genellikle 1.5*IQR'den daha uzakta olan değerleri gösterir.

# Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

**Ölüm şekline göre ayrılmış her iki cinsiyetteki yaşı swarm plot ile görselleştirin.**

**Swarmplot,** kategorik bir değişkenle sayısal bir değişken arasındaki ilişkiyi göstermek için kullanılır. 

Veri noktalarını bir arı sürüsü gibi kategorilerin üzerine yerleştirir, böylece her bir veri noktasını ayrı ayrı görmemizi sağlar. 

Bu, veri setindeki dağılımı ve yoğunluğu daha ayrıntılı bir şekilde anlamamıza yardımcı olur.

1. **Yoğunluk:**
   Veri noktalarının yoğun olduğu yerler, daha fazla noktanın bir arada olduğu yerlerdir.
   
2. **Dağılım:**
   Noktaların yerleşim şekli, veri setindeki dağılımı gösterir.
   
3. **Aykırı Değerler:**
   Diğerlerinden uzakta yer alan noktalar genellikle aykırı değerleri temsil eder.

Swarmplot, veri setindeki bireysel veri noktalarını gösterdiği için, genellikle veri setindeki varyansı, veri yoğunluğunu ve aykırı değerleri görmek için kullanılır. 

Bu plot, aynı zamanda, farklı kategoriler arasındaki dağılım farklılıklarını gözlemlemek için de kullanışlıdır.

```python
plt.figure(figsize = (10,6))
ax = sns.swarmplot(x="gender", y="age", hue="manner_of_death",data=kill)
```

# Plot 12 most dangerous cities

**En tehlikeli 12 şehri görselleştirin**

**Barplot (çubuk grafiği),** kategorik veri ile sayısal veri arasındaki ilişkiyi görselleştirmek için kullanılır. 

Barplot, belirli kategorilere ait sayısal değerleri temsil eden çubuklar çizer. 

Çubukların yükseklikleri, kategorilere ait sayısal değerleri gösterir. 

1. **Çubuk Yüksekliği:**
   Her bir çubuğun yüksekliği, o kategori için sayısal değeri temsil eder.
   
2. **Çubuk Genişliği:**
   Çubuk genişliği genellikle sabittir ve sadece kategoriyi temsil eder.
   
3. **Hata Çubukları:**
   Barplot'ta çubukların üstünde yer alan çizgiler (hata çubukları), güven aralığını gösterir.

Barplot, kategorik değişkenlerin sayısal değerlerini karşılaştırmak için kullanılır. 

Örneğin, farklı ürün kategorilerinin satış miktarlarını, farklı grupların ortalama puanlarını vb. karşılaştırmak için barplot kullanabilirsiniz. 

Barplot aynı zamanda, hata çubukları sayesinde, sayısal değerlerin güven aralıklarını da gösterir.

```python
ctyy= kill.city.value_counts().head(12)
ctyy
```

```python
yvalue= ctyy.values
yvalue
```

```python
xindex = ctyy.index
xindex
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax= sns.barplot(x=xindex, y= yvalue)

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3)
    
ax.set_xticklabels(ctyy.index, rotation=45, color="b")
ax.set_title("Most Dangerous Cities", fontsize=14, color="r")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(7,6))
ax = sns.countplot(x='city', data=kill, order= xindex)

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3)
    
ax.set_xticklabels(ctyy.index, rotation=45)
ax.set_title("Most Dangerous Cityes", fontsize=14, color="b")

plt.tight_layout()
```

# Plot Having mental ilness or not for killed people
**Öldürülen insanlar için akıl hastası olup olmama durumunu görselleştirin**

```python
kill.head(2)
```

```python
kill.signs_of_mental_illness.value_counts()
```

**with seaborn barplot**

```python
ax= sns.barplot(y="signs_of_mental_illness",  data=kill, errcolor="red")  # ci=None
```

sns.barplot fonksiyonu ile oluşturulan grafikteki üstündeki çizgi, çubuğun değerinin güven aralığını (confidence interval) gösterir. 

Güven aralığı, veri setindeki varyansı ve belirsizliği temsil eder. Eğer bir çubuğun güven aralığı genişse, bu, o kategorideki değerlerin daha dağılmış olduğunu gösterir. 

Güven aralığı çizgisini kaldırmak istiyorsanız, ci=None parametresini kullanabilirsiniz:

**countplot,** Seaborn kütüphanesinde bulunan bir fonksiyondur ve kategorik bir değişkenin sınıflarının frekansını saymak ve bu frekansları çubuklarla göstermek için kullanılır. Başka bir deyişle, countplot, belirli bir kategorik değişkenin her bir sınıfının kaç kez gözlemlendiğini gösterir.

1. **Çubuk Yüksekliği:**
   Her bir çubuğun yüksekliği, o kategori sınıfının sayısını (frekansını) temsil eder.
   
2. **Kategorik Eksen:**
   X veya Y ekseninde, gözlemlenen kategori sınıfları yer alır.
   
3. **Hata Çubukları:**
   Eğer kullanılırsa, güven aralığını veya başka bir hata ölçüsünü temsil eder.

countplot, kategorik değişkenlerin sınıf frekanslarını görmek ve bu sınıflar arasındaki dengeyi değerlendirmek için kullanılır. 

Örneğin, bir veri setindeki farklı renklerin, markaların, türlerin vb. frekanslarını göstermek için kullanılabilir.

**with seaborn countplot**

```python
fig, ax = plt.subplots(figsize=(8,6))

ax = sns.countplot(x='signs_of_mental_illness', data=kill)

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3)
```
