## Kütüphaneleri import edin

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python

import numpy as np
m = np.linspace(0,10,11)

print(f"result: \n\n{m}")
```

```python
c = 3 * 10**8 
E = m*c**2
print(f"Result: \n\n {E}")
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
import matplotlib.pyplot as plt

```

```python
plt.plot(m,E,color='b',lw=5)
plt.title("E=mc^2")
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.xlim(0,10)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(m,E,color='y',lw=5)
plt.title("E=mc^2")
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.xlim(0,10)
plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
plt.plot(m,E,color='g',lw=3)
plt.title("E=mc^2")
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.xlim(0,10)

# LOG SCALE
plt.yscale("log")
plt.grid(which='both',axis='y')

plt.show()
```

```python
#Bu iki kod satırı, logaritmik y eksenli bir çizgi grafiği çizmek ve yatay bir ızgara eklemek için kullanılır. 
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(labels, july16_2007,label='july16_2007')
axes.plot(labels,july16_2020,label='july16_2020')
plt.legend()

plt.show()
```

```python
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(labels, july16_2007,label='july16_2007')
axes.plot(labels,july16_2020,label='july16_2020')
plt.legend(loc=(1.04,0.5)) # location ile legend grafigin dışına aldık

plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig,axes = plt.subplots(2,1,figsize=(12,8))

axes[0].plot(labels, july16_2007,label='july16_2007')
axes[0].set_title("July 16th, 2007")
axes[1].plot(labels,july16_2020,label='july16_2020')
axes[1].set_title("July 16th, 2020")
#Her yılın getiri eğrisini gösteren grafik

plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, Ilk_x = plt.subplots(figsize=(12,8))

Ilk_x.plot(labels,july16_2007, lw=2, color="blue")
Ilk_x.set_ylabel("2007", fontsize=15, color="blue")

Ilk_x.spines['left'].set_color('blue')
Ilk_x.spines['left'].set_linewidth(4)

for label in Ilk_x.get_yticklabels():
    label.set_color("blue")
plt.yticks(fontsize=15)    
    
Ikinci_x = Ilk_x.twinx()
Ikinci_x.plot(labels,july16_2020, lw=2, color="red")
Ikinci_x.set_ylabel("2020", fontsize=15, color="red")

Ikinci_x.spines['right'].set_color('red')
Ikinci_x.spines['right'].set_linewidth(4)

for label in Ikinci_x.get_yticklabels():
    label.set_color("red")
    
Ilk_x.set_title("Years");
plt.yticks(fontsize=15);
```
