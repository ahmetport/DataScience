# Kütüphaneleri import edin

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Değişkenlerimizi tanımlayalım

**m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım**

Bu kod, 0 ile 10 arasında, her biri aralarında eşit mesafede olan 11 sayı üretir. 

Bu fonksiyon genellikle grafik çizimlerinde eksen değerleri oluşturmak, sayısal analiz yapmak veya belirli bir aralıktaki sayıları elde etmek için kullanılır.

```python
m = np.linspace(0, 10, 11)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

**PLOTTING E=mc^2**

- plt.plot(m, E, color = "blue", lw=3): m ve E değerlerini kullanarak bir çizgi grafik çizer. Burada color parametresi çizginin rengini blue olarak belirler ve lw parametresi çizginin kalınlığını 3 olarak ayarlar.
- plt.title("E = mc^2"): Grafik başlığını "E = mc^2" olarak belirler.
- plt.xlabel("M in g") ve plt.ylabel("E in J"): Sırasıyla x ve y eksenleri için etiket belirler.
- plt.xlim(0,10): X ekseni için değer aralığını 0 ile 10 arasında sınırlar.

```python
plt.plot(m, E, color = "blue", lw=3);
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0,10);
```

**figsize ve renk ile oynayalım**

- plt.figure(figsize= (8,6)): 8x6 boyutlarında yeni bir çizim penceresi açar.
- plt.plot(m, E, color = "orange", lw=3): m ve E değerlerini kullanarak bir çizgi grafik çizer. Burada color parametresi çizginin rengini orange olarak belirler ve lw parametresi çizginin kalınlığını 3 olarak ayarlar.
- plt.title("E = mc^2"): Grafik başlığını "E = mc^2" olarak belirler.
- plt.xlabel("M in g") ve plt.ylabel("E in J"): Sırasıyla x ve y eksenleri için etiket belirler.

```python
plt.figure(figsize= (6,4))

plt.plot(m, E, "bo", lw=5);
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J");
```

- fig = plt.figure(): Yeni bir çizim penceresi (figure) açar.
- ax = fig.add_axes([0, 0, 1, 1]): Çizim penceresine, pencerenin tamamını kaplayan yeni bir eksen (axes) ekler. [0, 0, 1, 1] koordinatları eksenin sol alt köşesini ve genişlik-yükseklik oranını belirtir.
- ax.plot(m, E, c="orange", lw=3): m ve E değerlerini kullanarak, çizgi rengi orange ve çizgi kalınlığı 3 olan bir çizgi grafik çizer.

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(m, E, c="orange", lw=2);
```

# Alttaki görseli oluştururken farklı bir değişken tanımlamayın. 

**Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)**

**Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir**

- plt.plot(m, E, color = "green", lw=3): m ve E değerlerini kullanarak bir çizgi grafik çizer. Burada color parametresi çizginin rengini green olarak belirler ve lw parametresi çizginin kalınlığını 3 olarak ayarlar.
- plt.title("E = mc^2"): Grafik başlığını "E = mc^2" olarak belirler.
- plt.xlabel("M in g") ve plt.ylabel("E in J"): Sırasıyla x ve y eksenleri için etiket belirler.
- plt.xlim(0,10): X ekseni için değer aralığını 0 ile 10 arasında sınırlar.
- plt.yscale("log"): Y ekseni için logaritmik ölçek kullanır.
- plt.grid(which="both", axis="y"): Y ekseni üzerinde, hem ana (major) hem de yardımcı (minor) çizgiler için ızgara çizgileri çizer.

```python
plt.plot(m, E, color = "green", lw=3);
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0,10)
plt.yscale("log");
plt.grid(which="both", axis="y")
```

# Aşağıdaki değişkenlere göre ilgili plotlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

- plt.figure(figsize=(7, 5)): 7x5 boyutlarında yeni bir çizim penceresi açar.
- plt.plot(labels, july16_2007, label = "july16_2007") ve plt.plot(labels, july16_2020, label = "july16_2020"): İki farklı veri serisini aynı grafik üzerine çizer ve her bir çizgi için etiket ekler.
- plt.legend(loc=5): Grafikteki çizgiler için etiketleri sağ orta (right center) konumda gösterir (loc=5 bu konumu belirtir).

```python
plt.figure(figsize=(7, 5))

plt.plot(labels, july16_2007, label = "july16_2007")
plt.plot(labels, july16_2020, label = "july16_2020")
plt.legend(loc=5);
```

- fig = plt.figure(): Yeni bir çizim penceresi (figure) açar.
- axes = fig.add_axes([0,0,1,1]): Çizim penceresine, pencerenin tamamını kaplayan yeni bir eksen (axes) ekler. [0,0,1,1] koordinatları eksenin sol alt köşesini ve genişlik-yükseklik oranını belirtir.
- axes.plot(labels, july16_2007, label = "july16_2007") ve axes.plot(labels, july16_2020, label = "july16_2020"): İki farklı veri serisini aynı grafik üzerine çizer ve her bir çizgi için etiket ekler.
- axes.legend(): Grafikteki çizgiler için etiketleri belirttiğiniz konumda gösterir.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(labels, july16_2007, label = "july16_2007")
axes.plot(labels, july16_2020, label = "july16_2020");
axes.legend();
```

- plt.figure(figsize=(8, 6)): 8x6 boyutlarında yeni bir çizim penceresi açar.
- plt.plot(labels, july16_2007, label = "july16_2007") ve plt.plot(labels, july16_2020, label = "july16_2020"): İki farklı veri serisini aynı grafik üzerine çizer ve her bir çizgi için etiket ekler.
- plt.legend(loc=(1.01, 0.4)): Grafikteki çizgiler için etiketleri çizimin hemen sağında, y ekseni üzerinden 0.4 konumunda gösterir.

```python
plt.figure(figsize=(8, 6))

plt.plot(labels, july16_2007, label = "july16_2007")
plt.plot(labels, july16_2020, label = "july16_2020")
plt.legend(loc=(1.01, 0.2));
```

- fig = plt.figure(): Yeni bir çizim penceresi (figure) açar.
- axes = fig.add_axes([0,0,1,1]): Çizim penceresine, pencerenin tamamını kaplayan yeni bir eksen (axes) ekler. [0,0,1,1] koordinatları eksenin sol alt köşesini ve genişlik-yükseklik oranını belirtir.
- axes.plot(labels, july16_2007, lw = 3, label = "july16_2007") ve axes.plot(labels, july16_2020, lw = 3, label = "july16_2020"): İki farklı veri serisini aynı grafik üzerine çizer, çizgi kalınlığını 3 yapar ve her bir çizgi için etiket ekler.
- axes.legend(loc=(1.01, 0.4)): Grafikteki çizgiler için etiketleri çizimin hemen sağında, y ekseni üzerinden 0.4 konumunda gösterir.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(labels, july16_2007, lw = 3, label = "july16_2007")
axes.plot(labels, july16_2020, lw = 3, label = "july16_2020");
axes.legend(loc=(1.01, 0.4));
```

# Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

**Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur.**

**ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.**

- fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (12,8)): 2 satır ve 1 sütundan oluşan, toplamda 2 alt grafik içeren, 12x8 boyutlarında bir çizim penceresi açar.
- axes[0].plot(labels, july16_2007, lw = 3, label = "july16_2007") ve axes[1].plot(labels, july16_2020, lw = 3, label = "july16_2020"): Sırasıyla ilk ve ikinci alt grafiklere, belirlenen çizgi kalınlığı ve etiket ile veri serilerini çizer.
- axes[0].set_title("july16_2007") ve axes[1].set_title("july16_2020"): Her bir alt grafiğe uygun başlıkları atar.

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (12,8))
axes[0].plot(labels, july16_2007, lw = 1, label = "july16_2007")
axes[0].set_title("july16_2007")
axes[1].plot(labels, july16_2020, lw = 3, label = "july16_2020")
axes[1].set_title("july16_2020");
```

# Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(labels, july16_2007, lw=3, color = "blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")


ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red");
```

- fig, ax1 = plt.subplots(figsize=(12,8)): 12x8 boyutlarında yeni bir çizim penceresi açar ve bir adet y ekseni (ax1) oluşturur.
- ax1.plot(labels, july16_2007, lw=3, color = "blue"): ax1 y ekseni üzerine, july16_2007 veri serisini çizer ve çizgi kalınlığını ve rengini belirler.
- ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue"): ax1 y ekseni için etiket, font büyüklüğü ve rengini belirler.
- ax2 = ax1.twinx(): ax1 ile aynı x ekseni üzerinde, farklı bir y ekseni (ax2) oluşturur.
- ax2.plot(labels, july16_2020, lw=3, color = "red"): ax2 y ekseni üzerine, july16_2020 veri serisini çizer ve çizgi kalınlığını ve rengini belirler.
- ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red"): ax2 y ekseni için etiket, font büyüklüğü ve rengini belirler.

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(labels, july16_2007, lw=3, color = "blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

1. fig, ax1 = plt.subplots(figsize=(12,8)): Bu satır, bir çizim penceresi (figure) ve bir eksen (axes) oluşturur. Çizim penceresinin boyutu figsize parametresi ile 12 birim genişlik ve 8 birim yükseklik olarak ayarlanır. Aynı zamanda, bu eksen ax1 adını alır.

2. ax1.spines["left"].set_color("blue") ve ax1.spines["left"].set_linewidth(2): Bu satırlar, ax1 ekseninin sol tarafındaki çizginin rengini mavi olarak ve kalınlığını 2 birim olarak ayarlar.

3. ax1.spines["right"].set_color("red") ve ax1.spines["right"].set_linewidth(2): Bu satırlar, ax1 ekseninin sağ tarafındaki çizginin rengini kırmızı olarak ve kalınlığını 2 birim olarak ayarlar.

4. ax1.plot(labels, july16_2007, lw=3, color="blue"): Bu satır, ax1 ekseninde labels ve july16_2007 veri serisini kullanarak bir çizgi grafiği çizer. Çizgi kalınlığı lw parametresi ile 3 birim olarak ve rengi "blue" (mavi) olarak belirlenir.

5. ax1.set_ylabel("July 16 2007", fontsize=20, color="blue"): Bu satır, ax1 eksenine bir y ekseni etiketi ekler. Etiketin metni "July 16 2007" olur, yazı boyutu 20 birim olur ve rengi mavi olur.

6. ax2 = ax1.twinx(): Bu satır, ax1 ile aynı x ekseni üzerinde farklı bir y ekseni olan ax2'yi oluşturur. twinx() işlevi, aynı x ekseni üzerinde farklı y ekseniyle çalışmanıza olanak tanır.

7. ax2.plot(labels, july16_2020, lw=3, color="red"): Bu satır, ax2 y ekseni üzerinde labels ve july16_2020 veri serisini kullanarak bir çizgi grafiği çizer. Çizgi kalınlığı lw parametresi ile 3 birim olarak ve rengi "red" (kırmızı) olarak belirlenir.

8. ax2.set_ylabel("July 16 2020", fontsize=20, color="red"): Bu satır, ax2 y ekseni için bir y ekseni etiketi ekler. Etiketin metni "July 16 2020" olur, yazı boyutu 20 birim olur ve rengi kırmızı olur.

Sonuç olarak, bu kod parçası iki farklı veri setini aynı grafik üzerinde göstermek için iki farklı renkte ve eksenlerle çalışarak kullanılır. Sol taraftaki eksen "July 16 2007" verisini temsil ederken, sağ taraftaki eksen "July 16 2020" verisini temsil eder. Bu, iki veri setini karşılaştırmak için kullanışlıdır.

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.spines["left"].set_color("blue")
ax1.spines["left"].set_linewidth(2)

ax1.spines["right"].set_color("red")
ax1.spines["right"].set_linewidth(2)

ax1.plot(labels, july16_2007, lw=3, color = "blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")


ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red");
```

1. fig, ax1 = plt.subplots(figsize=(12,8)): Bu satır, yeni bir çizim penceresi (figure) ve birinci eksen (ax1) oluşturur. Çizim penceresinin boyutu 12 birim genişlik ve 8 birim yükseklik olarak ayarlanır.

2. ax1.spines["left"].set_color("blue") ve ax1.spines["left"].set_linewidth(2): Bu satırlar, ax1'in sol y ekseni çizgisinin rengini mavi (blue) olarak ve kalınlığını 2 birim olarak ayarlar.

3. ax1.spines["right"].set_color("red") ve ax1.spines["right"].set_linewidth(2): Bu satırlar, ax1'in sağ y ekseni çizgisinin rengini kırmızı (red) olarak ve kalınlığını 2 birim olarak ayarlar.

4. ax1.plot(labels, july16_2007, lw=3, color="blue"): Bu satır, ax1 üzerine labels ve july16_2007 veri serisini kullanarak bir çizgi grafiği çizer. Çizgi kalınlığı lw parametresi ile 3 birim olarak ayarlanır ve rengi mavi (blue) olarak belirlenir.

5. ax1.set_ylabel("July 16 2007", fontsize=20, color="blue"): Bu satır, ax1'in y ekseni için bir etiket (label) belirler. Etiketin metni "July 16 2007" olur, yazı boyutu (fontsize) 20 birim olarak ayarlanır ve rengi mavi (blue) olur.

6. for label in ax1.get_yticklabels(): label.set_color("blue"): Bu döngü, ax1'in y ekseninin etiketlerinin rengini mavi (blue) olarak ayarlar. Yani, y eksenindeki sayılar mavi renkte görünecektir.

7. ax2 = ax1.twinx(): Bu satır, ax1 ile aynı x ekseni üzerinde ikinci bir y ekseni (ax2) oluşturur. twinx() işlevi, aynı x ekseni üzerinde farklı y ekseniyle çalışmanıza olanak tanır.

8. ax2.plot(labels, july16_2020, lw=3, color="red"): Bu satır, ax2 y ekseni üzerine labels ve july16_2020 veri serisini kullanarak bir çizgi grafiği çizer. Çizgi kalınlığı lw parametresi ile 3 birim olarak ayarlanır ve rengi kırmızı (red) olarak belirlenir.

9. ax2.set_ylabel("July 16 2020", fontsize=20, color="red"): Bu satır, ax2 y ekseni için bir etiket (label) belirler. Etiketin metni "July 16 2020" olur, yazı boyutu (fontsize) 20 birim olarak ayarlanır ve rengi kırmızı (red) olur.

10. for label in ax2.get_yticklabels(): label.set_color("red"): Bu döngü, ax2'nin y ekseninin etiketlerinin rengini kırmızı (red) olarak ayarlar. Yani, ikinci y eksendeki sayılar kırmızı renkte görünecektir.

Bu kod parçası, iki farklı veri setini aynı grafikte göstermek için iki farklı renkte ve eksenlerle çalışarak kullanılır. Sol taraftaki eksen "July 16 2007" verisini temsil ederken, sağ taraftaki eksen "July 16 2020" verisini temsil eder. Bu, iki veri setini karşılaştırmak veya ilişkilendirmek için kullanışlıdır.

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.spines["left"].set_color("blue")
ax1.spines["left"].set_linewidth(2)

ax1.spines["right"].set_color("red")
ax1.spines["right"].set_linewidth(2)

ax1.plot(labels, july16_2007, lw=3, color = "blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```
